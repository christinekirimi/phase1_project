```python
#Checking for Null values the dataframe in the original title column
#There are 21 missing values in the original title column
title_basics_df[title_basics_df["original_title"].isna()]
```

```python
#Checking for Null values the dataframe in the runtime_minutes column
#There are 31739 missing values in the runtime_minutes column
title_basics_df[title_basics_df["runtime_minutes"].isna()]
```

```python
#Checking for Null values the dataframe in the genres column
#There are 5408  missing values in the genres column
title_basics_df[title_basics_df["genres"].isna()]
```

```python
#Cleaning the foeign gross column: removing "," and "." using the replace method
bom_movie_df["foreign_gross"] = bom_movie_df["foreign_gross"].str.replace(',', '')
bom_movie_df["foreign_gross"] = bom_movie_df["foreign_gross"].str.replace('.', '')
bom_movie_df["foreign_gross"] = bom_movie_df["foreign_gross"].astype("float")
bom_movie_df

```

```python
# calculating the mean and median for the domestic gross column
dg_median = bom_movie_df["domestic_gross"].median()
print(dg_median)
dg_mean = bom_movie_df["domestic_gross"].mean()
print(dg_mean)
```

```python
#Replacing the misssing values in the domestic gross column with the median
bom_movie_df["domestic_gross"] = bom_movie_df["domestic_gross"].fillna(bom_movie_df["domestic_gross"].dropna().median())
bom_movie_df
```

```python
# calculating the mean and median for the foreign gross column
fg_median = bom_movie_df["foreign_gross"].dropna().median()
print(fg_median)
fg_mean = bom_movie_df["foreign_gross"].dropna().mean()
print(fg_mean)
```

```python
#Replacing the misssing values in the foreign gross column with the median
bom_movie_df["foreign_gross"] = bom_movie_df["foreign_gross"].fillna(bom_movie_df["foreign_gross"].dropna().median())
bom_movie_df
```

```python
#checking the value_counts
bom_movie_df["studio"].value_counts()
```

```python
#replacing the null studio with the most common movie studio "IFC"(studio with the highest Counts)
bom_movie_df["studio"] = bom_movie_df["studio"].fillna("IFC")
bom_movie_df["studio"].value_counts()
```

```python
#Replacing the missing Original Title with the Primary title of the movie
title_basics_df["original_title"] = title_basics_df["original_title"].fillna(title_basics_df["primary_title"])
title_basics_df["original_title"]
```

```python
#checking for the median of runtime minutes
title_basics_df["runtime_minutes"].dropna().median()
```

```python
#checking for the mean of runtime minutes
title_basics_df["runtime_minutes"].dropna().mean()
```

```python
# replacing the missing runtime values with the median
title_basics_df["runtime_minutes"] = title_basics_df["runtime_minutes"].fillna(title_basics_df["runtime_minutes"].dropna().median())
title_basics_df["runtime_minutes"]
```

```python
# removing the rows in the genres column that have missing genres
title_basics_df = title_basics_df.dropna(subset =["genres"])
title_basics_df["genres"]

```

```python
#checking the title basics dataframe to ensure that it is clean
title_basics_df
```

## DATA ANALYSIS

```python
bom_movie_df.describe()
```

Answer what is the distribution of the movies by studio?

```python

# Set up plots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 5))

# Create variables for easier reuse
value_counts = bom_movie_df["studio"].value_counts()
top_10_counts = value_counts.iloc[:10]

# Plot data
ax1.bar(value_counts.index, value_counts.values)
ax2.bar(top_10_counts.index, top_10_counts.values)

# Customize appearance
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=45)
ax1.set_ylabel("Count of movies")
ax2.set_ylabel("Count of movies")
ax1.set_title("Distribution of movies by studio")
ax2.set_title("Top 10 studios by Count of movies");
```

How many movies were produced in Each year?

```python

# checking out the distribution of movies per year
x = bom_movie_df.groupby("year")["title"].count()
print(x)
x.plot(kind="bar",color="blue",edgecolor="black")

plt.xlabel('Year')
plt.ylabel('No of movies')
plt.title("Distribution of movies per year")
plt.show()
```

```python
#Checking out the distribution of amount of revenue earned per year
Domestic_Revenue_Year =bom_movie_df.groupby("year")["domestic_gross"].sum()
print(Domestic_Revenue_Year)
plt.figure(figsize=(10,6))
Domestic_Revenue_Year.plot(kind="bar",color="blue",edgecolor="black")
#Chart title and Labels
plt.title("Domestic Revenue by Year")
plt.xlabel("Year")
plt.ylabel("Domestic Revenue")
plt.show
```

```python
#Checking out the distribution of amount of revenue earned per year
Foreign_Revenue_Year =bom_movie_df.groupby("year")["foreign_gross"].sum()
print(Foreign_Revenue_Year)
plt.figure(figsize=(10,6))
Foreign_Revenue_Year.plot(kind="bar",color="blue",edgecolor="black")
#Chart title and Labels
plt.title("Foreign Revenue by Year")
plt.xlabel("Year")
plt.ylabel("Foreign Revenue")
plt.show
```

```python
#Checking the correlation between Domestic Gross and Foreign Gross
x = bom_movie_df["domestic_gross"]
y = bom_movie_df["foreign_gross"]
# Pass in x and y values with a label 
plt.scatter(x, y, label = "correlation between domestic gross and Foreign gross" )
plt.title('Relationship Between Domestic Gross and Foreign Gross')
plt.legend()
```

### COMBINING TITLE BASICS AND TITLE RATINGS DATA FRAMES

```python
#performing an outer merge.
Basics_ratings_df = pd.merge(title_basics_df,title_ratings_df, how = "outer", on ="tconst")
Basics_ratings_df.info()
```

```python
Basics_ratings_df = Basics_ratings_df.dropna()
Basics_ratings_df
```

```python
 Basics_ratings_df.describe()
```

```python
plt.figure(figsize=(10,6))
genre_runtime = Basics_ratings_df.groupby("genres")["runtime_minutes"].mean().nlargest(20, "last")
print(genre_runtime)
genre_runtime.plot(kind="bar", color="blue", edgecolor="black")

#Chart title and Labels
plt.title("Average runtime by movie genre")
plt.xlabel("Genre")
plt.ylabel("Runtime")
plt.show
```

```python
genre_rating = Basics_ratings_df.groupby("genres")["averagerating"].mean().nlargest(20)
#print(genre_rating)
plt.figure(figsize=(10,6))
genre_rating.plot(kind="bar",color="blue",edgecolor="black")
#Chart title and Labels
plt.title("average rating per genre")
plt.xlabel("Genre")
plt.ylabel("average rating")
```

```python
genre_votes = Basics_ratings_df.groupby("genres")["numvotes"].sum().nlargest(20)
#print(genre_votes)
plt.figure(figsize=(10,6))
genre_votes.plot(kind="bar",color="blue",edgecolor="black")
#Chart title and Labels
plt.title("number of votes per genre")
plt.xlabel("Genre")
plt.ylabel("number of votes")
plt.show
```

correlation between average rating and number of votes

```python
x = title_ratings_df["averagerating"]
y = title_ratings_df["numvotes"]
# Pass in x and y values with a label 
plt.scatter(x, y, label = "correlation between number of votes and Average Rating" )
plt.title('Relationship Between number of votes and Average Rating')
plt.legend()
plt.show()
```

## Merge The Basic Ratings Data Frame With The Bom Movie Data Frame

```python
movie_df = pd.merge(Basics_ratings_df,bom_movie_df, left_on=("primary_title"), right_on =("title"))
movie_df
```

```python
#getting a new column_totalgross by combining domestic gross and foreign_gross
movie_df["total_gross"] = movie_df["domestic_gross"] + movie_df["foreign_gross"]
movie_df
```

```python
#visualizing th relationship between total revenue i.e total gross and the genres
plt.figure(figsize=(10,6))
genre_revenue = movie_df.groupby("genres")["total_gross"].sum().nlargest(20, "all")
genre_revenue.plot(kind="bar", color="blue", edgecolor="black")

#Chart title and Labels
plt.title("Revenue by movie genre")
plt.xlabel("Genre")
plt.ylabel("Revenue")
plt.show
```

```python
#visualizing th relationship between total revenue i.e total gross and the runtime
plt.figure(figsize=(10,6))
runtime_revenue = movie_df.groupby("runtime_minutes")["total_gross"].sum().nlargest(20, "all")
print(runtime_revenue)
runtime_revenue.plot(kind="bar", color="blue", edgecolor="black")

#Chart title and Labels
plt.title("Revenue by runtime")
plt.xlabel("Runtime")
plt.ylabel("Revenue")
plt.show
```
